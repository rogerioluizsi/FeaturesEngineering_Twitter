# Statistics Tests - Some backgroud hypothesis

```python
#independent samples: 
bot = database[database['Alvo']==1]  
gen = database[database['Alvo']==0]  
```

```python
#selectig features to the test
columns = ["n_retweet_mean", "n_mentions_mean", "in_tweets_reply", 'age', 'size_name', 'size_sname', 'statuses_count', "n_number_sname"]
bot[columns].describe()
```

```python
#those samples of class Genuine does not follow normal distribution. p-value <0.05
for i in columns:
    stat, p = stats.normaltest(gen[i])
    print(round(p, 4), i)
```

```python
#those samples of class Bot does not follow normal distribution. p-value <0.05
for i in columns:
    stat, p = stats.normaltest(bot[i])
    print(round(p, 4), i)
```

 let's use a non-parametric median test

## 1 - Hypothesis
Do Genuine Accouts are more retweeted?

H0: Bots retweet like a genuine accounts

H1: Bots retweet more than genuine accounts

```python
print("Bot retweet mean",bot['n_retweet_mean'].mean())
print("Genuine retweet mean",gen['n_retweet_mean'].mean())
stats.mannwhitneyu(bot['n_retweet_mean'].values, gen['n_retweet_mean'].values, alternative='less')
```

```python
print("Bot retweet mean",bot['statuses_count'].mean())
print("Genuine retweet mean",gen['statuses_count'].mean())
stats.mannwhitneyu(bot['statuses_count'].values, gen['statuses_count'].values, alternative='greater')
```

##### :) low P-value reject the null hypothesis. Than, i have statical evidences to accept tha genuine account are more retweeted than bots
